```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

def display_image(image_url, title=None):
    try:
        response = requests.get(image_url)
        img = Image.open(BytesIO(response.content))
        plt.figure(figsize=(5, 5))
        if title:
            plt.title(title, fontsize=16)
        plt.imshow(img)
        plt.axis('off')
        plt.show()
    except Exception as e:
        print(f"Could not load image from {image_url}: {e}")
```

### Write into file

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import json

json_entries = []

for idx in range(0, len(meta_dataset)):  # Adjust range as needed
    image_url = None
    for url in meta_dataset[idx]["images"]["hi_res"]:
        if url is not None:
            image_url = url
            break
    
    if image_url is None:
        continue
    
    title = meta_dataset[idx]["title"]
    parent_asin = meta_dataset[idx]["parent_asin"]
    
    # Create JSON entries
    index_entry = {"index": {"_index": "images", "_id": str(parent_asin)}}
    data_entry = {
        "category_ids": [62],
        "url": image_url,
        "caption": title,
        "source": "amazon"
    }
    
    json_entries.append(index_entry)
    json_entries.append(data_entry)

# Write the JSON entries to a file
with open('amazon_es.json', 'w') as f:
    for entry in json_entries:
        f.write(json.dumps(entry) + "\n")
```
